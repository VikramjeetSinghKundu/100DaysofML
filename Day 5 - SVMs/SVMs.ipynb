# Let's start with our model training

```python
from sklearn.model_selection import train_test_split
```

```python
# Spcifying the features and the target variable

X = df.drop(['target', 'flower_name'], axis='columns')
y = df['target']
```

```python
#Performing the train, test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
#Initializing the model

from sklearn.svm import SVC

model = SVC()
```

```python
#Fitting it on the training set

model.fit(X_train, y_train)
```

```python
#Making predicition on the test set. Using features in the test set (X_test) to predict values for y_test

y_pred = model.predict(X_test)
y_pred
```

```python
# Evaluating accuracy of the trained model on the test set. 
# Comparing the predicted values (y_pred) to the actual values (y_test) for an accuracy score,

#The predictions (y_pred) are compared to the true labels (y_test) to determine how many are correct.
model.score(X_test,y_test)
```
