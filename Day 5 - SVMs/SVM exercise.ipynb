```python
#Importing libraries and dataset

import pandas as pd
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
dir(digits)
```

```python
digits['target_names']
```

```python
digits['data']
```

```python
#Converting the loaded dataset into a dataframe for ease of use

df = pd.DataFrame(digits['data'], digits['target'])
```

```python
df
```

```python
df['target'] = digits['target']
```

```python
df
```

```python
# Specifying the dependent (target) and independent variables (features)

X = df.drop(['target'], axis='columns')
y = df['target']
```

# Performing the train_test_split


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
#Initializing the rbf model first

from sklearn.svm import SVC

rbf_model = SVC(kernel = 'rbf')
```

```python
rbf_model.fit(X_train, y_train)
```

```python
y_pred_rbf = rbf_model.predict(X_test)
y_pred_rbf
```

```python
rbf_model.score(X_test, y_test)
```

```python
# A more inuitive way of calculating accuracy

from sklearn.metrics import accuracy_score
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
accuracy_rbf
```

# Using linear kernel

```python
linear_model = SVC(kernel ='linear')
linear_model.fit(X_train, y_train)
```

```python
linear_model.score(X_test, y_test)
```

```python
y_pred_linear = linear_model.predict(X_test)
y_pred_linear
```

```python
# A more inuitive way of calculating accuracy
accuracy_linear = accuracy_score(y_test, y_pred_linear)
accuracy_linear
```
