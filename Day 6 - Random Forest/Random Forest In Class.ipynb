```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
import matplotlib.pyplot as plt

plt.gray()
for i in range(10):
    plt.matshow(digits.images[i])
    
# 8*8 pixels of handwritten images
```

```python
digits['data'][:5] #each number/element is a 1d array (8*8 matrix) and the length is therefore 64 pixels
```

```python
#converting this data into a dataframe for ease of use
df = pd.DataFrame(digits['data'])
df  #each sample is nothing but an array of 64 integers (which form a handwritten number) and they map to 
    #one of the target variable
```

```python
 digits['target']
```

```python
#appending the target variable into the dataframe 

df['target'] = digits['target']
df
```

# starting out with model creating and training

```python
#Specifiying the target variable and the features 

X = df.drop(['target'], axis = 'columns') #independent variables or features
y = df['target'] #target variable
```

```python
#splitting the dataset into train test splits

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

# Importing and initializing the model

```python
from sklearn.ensemble import RandomForestClassifier #ensemble is used when we are using multiple algorithm to predict the outome
model = RandomForestClassifier(n_estimators = 50)
print(model.get_params())
```

```python
#Training the model

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test) #n_estimators is the number of decision trees in the forest
```

```python
y_pred
```

```python
model.score(X_test, y_test)
```

# Better way to evaluate the model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predictions from the model
y_pred = model.predict(X_test)

# Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')  # Weighted for multiclass (0-9)
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

# Plotting the confusion matrix (truth on one axis and the predicted values on the other)

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
#Visualising it in a better way using seaborn

import seaborn as sns
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted values')
plt.ylabel('True values')
```
